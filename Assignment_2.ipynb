```python
# Detecting a file format of any kind
def get_file_format(file):
    
    with open(file, "rb") as f:
        file_content = f.read(10)

    try:  
        if file_content.startswith(b"%PDF"):
            return "PDF"
        elif file_content.startswith(b"\x89PNG"):
            return "PNG"
        elif file_content.startswith(b"\xff\xd8"):
            return "JPEG"
        elif file_content.startswith(b"GIF8"):
            return "GIF"
        else:
            with open(file, "r") as f:
                file_content = f.readline()
            if type(file_content) != list and file_content.startswith('{'):
                return 'JSON'
            elif type(file_content) != list and "," in file_start:
                return 'CSV'
            elif "<html" or '!' in file_content.lower():
                return "HTML"
            
    except Exception:
        f_extns = file.split(".")
        file_format = (f_extns[-1]).upper()
        return file_format
    
#/Users/varsha/Desktop/Sem 4/Stats TA/Stats TA 2/Exams/Exam 1/Seating - 5706.xlsx
file_path = input("\nEnter the path to the file: \n")
file_format = get_file_format(file_path)
print("\nThe file format is: {}".format(file_format))
```

```python
# Creating class "Read_Data"
class Read_Data:
    # Creating method "Get_Num_Lines" to get the number of lines to be printed
    def Get_Num_Lines(self):
        self.n_lines = int(input("\nEnter the number of lines: "))
        print()
    
    # Creating method "Print_Lines" to print the n number of lines
    def Print_Lines(self, file_path):
        with open(file_path, 'r') as f:
            lines = f.readlines()
            for text in (lines[:self.n_lines]):
                print(text, end ='')
        print()
        f.close()

# Accessing the class and methods
file_path = "/Users/varsha/Desktop/Sem 4/Web scraping/Assignment 2/Electric_Vehicle_Population_Data.json"        
rd = Read_Data()
rd.Get_Num_Lines()
rd.Print_Lines(file_path)
```

```python
# Extracting images from the URL
try:
    from urllib.parse import urlparse # Python 3
except ImportError:
    from urlparse import urlparse # Python 2

url = "https://en.wikipedia.org/wiki/Twitter"

response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Finding and extracting the image tags
img_tags = soup.find_all('img')
pprint(img_tags)
```

```python
# Extracting URLs from the image tags
img_urls = [img['src'] for img in img_tags]

# Getting the list of image file names
for i in img_urls:
    disassembled = urlparse(i)
    filename, file_ext = splitext(basename(disassembled.path))
    print (filename)
```

```python
# Displaying the last image on the website
url = urlopen("https://en.wikipedia.org" + img_tags[-1]['src']).read()
from IPython import display
display.Image(url)
```

```python
# Extracting the years from the website
soup = BeautifulSoup(response.content, 'html.parser')

years = []
four_digits = re.findall('[1-2][0-9][0-9][0-9]', str(soup))
for i in four_digits:
    if int(i)>=1950 and int(i)<=2050:
        years.append(i)
years
```

```python
# Getting all the lines that have the word "pandemic" in them
tags = []
for tag in soup.find_all('p'):
    tags.append(tag.text)

pandemic_content = []
for i in tags:
    if 'pandemic' in i:
        for line in i.split('.'):
            if 'pandemic' in line:
                pandemic_content.append(line)
for i in pandemic_content:
    print("\n" + i)

# Creating and writing the pandemic lines to a JSON file
with open('pandemic_content.json', 'w') as f:
    json.dump(pandemic_content, f)
```

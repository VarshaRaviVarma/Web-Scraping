```python
# Loading the HTML file
with open('/Users/varsha/Desktop/Sem 4/Web scraping/Assignment 4/NOPAGETAB_NFHLWMS.html', 'r') as file:
    contents = file.read()

soup = BeautifulSoup(contents, 'html.parser')
text= soup.get_text()

# Finding the number of times the word was mentioned in the file
# CASE SENSITIVE
word = "Hazard"
print ("\nCase Sensitive - The number of times the word Hazard shows up in the HTML file is: " + str(len(re.findall(word, text))))
# NOT CASE SENSITIVE
word = "hazard"
print ("\nNot case Sensitive - The number of times the word Hazard shows up in the HTML file is: " + str(len(re.findall(word, text.lower()))))
```

```python
# Extracting the time mentioned in the file
time = re.search(r'\d{1}:\d{1}\d{1} PM ET', text)
print ("\n" + str(time))
```

```python
# Extracting all URL addresses written in the file
pattern = re.compile('<a.*?href="(.*?)".*?>')
urls = re.findall(pattern, contents)

print ("\nThe number of URLs found in the file is: " + str(len(urls)) + "\n")
print(urls)
```
